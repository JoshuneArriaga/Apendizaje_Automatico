```python
import pandas as pd
```

```python

df = pd.read_csv("/content/newborn_health_monitoring_with_risk.csv")
df.head()
```

## **Tipo de columnas**

```python
df.dtypes
```

## **Crear, modificar o eliminar columnas**

```python
df_c = df.copy()
df_c = df.rename(columns={'name' : 'nombre'})
df_c
```

```python
df_c = df_c.drop('stool_count', axis=1)
df_c
```

```python
df_c['weight_change_kg'] = df_c['weight_kg'] - df_c['birth_weight_kg']
df_c.head()
```

## **Filtrar registros**

```python
at_risk_babies = df_c[df_c['risk_level'] == 'At Risk']
at_risk_babies.head()
```

```python
babies_older_than_10_days = df_c.query('age_days > 10')
babies_older_than_10_days.head()
```

## **Crear subconjutos de datos**

```python
subset_columns = df_c[['nombre', 'age_days', 'weight_kg', 'risk_level']]
subset_columns.head()
```

```python
subset_at_risk = df_c.loc[df_c.risk_level  == 'At Risk']
subset_at_risk.head()
```

## **Crear grupos de datos**

```python
grouped_by_risk = df_c.groupby('risk_level')
grouped_by_risk.head()
```

```python
risk_group_sizes = df_c.value_counts('risk_level')
risk_group_sizes
```

```python
average_weight_by_risk = grouped_by_risk['weight_kg'].mean()
average_weight_by_risk
```

```python
average_age_by_gender = df_c.groupby('gender')['age_days'].mean()
average_age_by_gender
```

```python
import pandas as pd
from scipy.stats import shapiro, skew, kurtosis
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/content/newborn_health_monitoring_with_risk.csv")
df
```

## Aplica algún método de filtro a tus datos mediante el uso de SelectKBest

```python
df['apgar_score'] = df.groupby('baby_id')['apgar_score'].ffill()
df['apgar_score'] = df.groupby('baby_id')['apgar_score'].bfill()
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
label = LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col] = label.fit_transform(df[col].astype(str))

df
```

```python
x = df.drop(columns=['risk_level','name'])
y = df['risk_level']
```

```python
escalador = MinMaxScaler()
escala = escalador.fit_transform(x)
x_escalada = pd.DataFrame(escala, columns=x.columns)
x_escalada
```

```python
selector = SelectKBest(score_func=f_classif, k=10)
X_new = selector.fit_transform(x_escalada, y)
```

```python
selected_features = x_escalada.columns[selector.get_support()]
print("Características seleccionadas con SelectKBest:")
print(selected_features)
```

## Aplicar métodos de selección basados en modelos

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
model_lr = LogisticRegression(max_iter=5000)
rfe = RFE(model_lr, n_features_to_select=10)
rfe.fit(x_escalada, y)

selected_rfe = x.columns[rfe.support_]
print("Características seleccionadas con RFE (Regresión Logística):")
print(selected_rfe)
```

## Métricas de selección de características



*   **Recursive Feature Elimination (RFE)**
La eliminación recursiva de características, o RFE por sus siglas en inglés, es un algoritmo de selección de características.

*   **Mutual Information**
La información mutua (IM) es un concepto fundamental de la teoría de la información que cuantifica la cantidad de información que dos variables aleatorias se proporcionan mutuamente. Mide la reducción de la incertidumbre de una variable gracias al conocimiento de la otra.
